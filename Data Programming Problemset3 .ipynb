```python
import warnings
warnings.filterwarnings("ignore")
# Step 1
import pandas as pd

# Step 2
URL = r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'

# Step 3
users = pd.read_csv(URL, sep='|')

# Step 4
print(users.groupby('occupation').age.mean())
```

```python
# Step 5
gender_records = users.pivot_table(index='occupation', columns='gender', aggfunc='size', fill_value=0)
sums= gender_records[['F', 'M']].sum(axis=1)
gender_records['Male Ratio'] = round(100 * gender_records['M'] / sums , 1)
gender_records['Male Ratio'].sort_values(ascending=False)
```

```python
# Step 6
users.groupby('occupation').age.agg(['min','max'])
```

```python
# Step 7
users.groupby(['occupation', 'gender']).agg({'age':['mean']})
```

```python
# Step 8
user_count=users.groupby(['occupation','gender'])['gender'].count()
total_Count=users.groupby(['occupation'])['gender'].count()
gender_percent=((user_count/total_Count)*100)
gender_percent
```

```python
# Step 1
import pandas as pd
import datetime

# Step 2
data = pd.read_csv("wind.txt")


# Step 3
data = pd.read_csv("wind.txt", sep = "\s+", parse_dates = {'Date':[0,1,2]})
data.head()

```

```python
# Step 4
import datetime as dt

def correct_year(x):
    if x.year > 2021:
        year = x.year - 100
    else:
        year = x.year
    return dt.date(year,x.month,x.day)

data['Date']=data['Date'].apply(lambda x: correct_year(x))
data
```

```python
# Step 5
data['Date'] = data['Date'].astype('datetime64[ns]')
data = data.set_index('Date')
```

```python
# Step 6
print(data.isnull().sum())

# Step 7
(data.notnull().sum())
print(data.notnull().sum().sum())


# Step 8
print(data.mean().mean())
```

```python
# Step 9
loc_stats = data.agg(["count","min","max","mean","std"])
loc_stats
```

```python
# Step 10
day_stats=pd.DataFrame()
day_stats['min']=data.min(axis=1)
day_stats['max']=data.max(axis=1)
day_stats['mean']=data.mean(axis=1)
day_stats['std']=data.std(axis=1)
day_stats.head()

# Step 11
avg_wind=data[data.index.month==1]
avg_wind.mean()
```

```python
# Step 12
yearly_frequency = data.resample(rule = "Y")
print(yearly_frequency.mean())
```

```python
# Step 13
yearly_frequency = data.resample(rule = "M")
print(yearly_frequency.mean())
```

```python
# Step 14
weekly_frequency = data.resample(rule = "W")
print(weekly_frequency.mean())
```

```python
# Step 15
weekly = data.resample(rule = "W").agg(['min','max','mean','std'])
print(weekly.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("us-marriages-divorces-1867-2014.csv")
plt.figure(figsize=(20,8))
plt.plot(df.Year,df.Marriages_per_1000,label="Marriages",color="green")
plt.plot(df.Year,df.Divorces_per_1000,label="Divorces",color="red")
plt.title('Marriages and Divorces in USA between 1867 and 2014',fontsize=18)
plt.xlabel('Year',fontsize=18,color="green")
plt.ylabel('Per Capita',fontsize=18,color="green")
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend()
plt.xticks(df.Year[::10])
plt.show()
```

# Question 7: 

```python
df = df[(df.Year == 1900) | (df.Year == 1950) |(df.Year == 2000)]
df.set_index('Year')[['Marriages_per_1000', 'Divorces_per_1000']].plot(kind='bar', figsize=(10, 5))
plt.xticks(rotation=360)
plt.title("Comparing number of marriages and divorces per capita in U.S. between 1900, 1950, and 2000", fontsize=16, y=1.01)
plt.xlabel("Years", labelpad=16,color="brown")
plt.ylabel("Per Capita.1000", labelpad=16,color="brown")
plt.legend(["Marriages", "Divorces"], fontsize=14)
```

```python
roman_Data= pd.read_csv('roman-emperor-reigns.csv')
roman_Data.head()
lower_Data= roman_Data.where(roman_Data.Cause_of_Death=="Assassinated").Cause_of_Death.count()
remaining_data= roman_Data.Cause_of_Death.count()-lower_Data
label=["Other Cause of Deaths","Assassinated"]
plt.title("Roman Emperors")
plt.pie([remaining_data,lower_Data],labels=label,autopct='%.4f%%',shadow = True, colors= ['Purple','Pink','Blue'])
plt.show()
```
